#### Initialisation - for only 1 galaxy z_i, F_i

```python
import sys
sys.path.append('../')
import numpy as np
from delight.io import *
from delight.utils import *
#from rail.estimation.algos.include_delightPZ.libPriorPZ import libPriorPZ
from delight.photoz_gp import PhotozGP
from delight.photoz_kernels import Photoz_mean_function, Photoz_kernel
from delight.utils_cy import approx_flux_likelihood_cy
import scipy.stats
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
import logging
import coloredlogs
import os
from time import time

# Plot formatting
plt.rcParams["figure.figsize"] = (20, 20)
plt.rcParams["axes.labelsize"] = 'xx-large'
plt.rcParams['axes.titlesize'] = 'xx-large'
plt.rcParams['xtick.labelsize']= 'xx-large'
plt.rcParams['ytick.labelsize']= 'xx-large'
```

```python
paramFileName = 'tmp/parametersTest_11.cfg'
params = parseParamFile(paramFileName, verbose=False)
```

```python
DL = approx_DL()

bandCoefAmplitudes, bandCoefPositions, bandCoefWidths, norms = readBandCoefficients(params)
numBands = bandCoefAmplitudes.shape[0]

redshiftDistGrid, redshiftGrid, redshiftGridGP = createGrids(params)
numZ = redshiftGrid.size

# Locate which columns of the catalog correspond to which bands.

bandIndices, bandNames, bandColumns, bandVarColumns, redshiftColumn,refBandColumn = readColumnPositions(params, prefix="target_")

dir_seds = params['templates_directory']
dir_filters = params['bands_directory']
lambdaRef = params['lambdaRef']
sed_names = params['templates_names']

# f_mod  : flux model in each band as a function of the sed and the band name
# axis 0 : redshifts
# axis 1 : sed names
# axis 2 : band names

f_mod = np.zeros((redshiftGrid.size, len(sed_names),len(params['bandNames'])))
print(f_mod.shape)
print(sed_names)
```

```python
# loop on SED to load the flux-redshift file from the training
# ture data or simulated by simulateWithSEDs.py

for t, sed_name in enumerate(sed_names):
    f_mod[:, t, :] = np.loadtxt(dir_seds + '/' + sed_name + '_fluxredshiftmod.txt')
print(f_mod[0, 0, :])
```

```python
numObjectsTraining = np.sum(1 for line in open(params['training_catFile']))
redshifts = np.zeros((numObjectsTraining, ))
model_mean = np.zeros((numZ, numObjectsTraining, numBands))
model_covar = np.zeros((numZ, numObjectsTraining, numBands))
bestTypes = np.zeros((numObjectsTraining, ), dtype=int)
ells = np.zeros((numObjectsTraining, ), dtype=int)

# loop on training data and training GP coefficients produced by delight_learn
# It fills the model_mean and model_covar predicted by GP
loc = -1
trainingDataIter = getDataFromFile(params, 0, numObjectsTraining, prefix="training_", ftype="gpparams")

# loop on training data to load the GP parameter
for loc, (z, ell, bands, X, B, flatarray) in enumerate(trainingDataIter):
    t1 = time()
    redshifts[loc] = z              # redshift of all training samples
    #gp.setCore(X, B, nt,flatarray[0:nt+B+B*(B+1)//2])
    #bestTypes[loc] = gp.bestType   # retrieve the best-type found by delight-learn
    ells[loc] = ell                # retrieve the luminosity parameter l

    # here is the model prediction of Gaussian Process for that particular trainning galaxy
    #model_mean[:, loc, :], model_covar[:, loc, :] = gp.predictAndInterpolate(redshiftGrid, ell=ell)
    t2 = time()
    #print(loc, t2-t1)
```

```python
numObjectsTarget = np.sum(1 for line in open(params['target_catFile']))
print("Number of targets : {}".format(numObjectsTarget))

firstLine = 0
lastLine = numObjectsTarget
numLines = lastLine - firstLine

# Now loop over each target galaxy (indexed bu loc index) to compute likelihood function
# with its flux in each bands
targetDataIter = getDataFromFile(params, firstLine, lastLine, prefix="target_", getXY=False)

ind, (z, normedRefFlux, bands, fluxes, fluxesVar, bCV, fCV, fvCV) = list(enumerate(targetDataIter))[0]
print(ind, z, bands)

print("f_mod is of dimensions {}".format(f_mod.shape))



fig, axs = plt.subplots(4, 2, constrained_layout=True)
fig2, axs2 = plt.subplots(4, 2, constrained_layout=True)
fig3, axs3 = plt.subplots(4, 2, constrained_layout=True)
fig4, axs4 = plt.subplots(4, 2, constrained_layout=True)

sigmaList=np.logspace(0, 15, 16)
ell_hat_z = 1
sigmaIter=-1

for ellPriorSigma in sigmaList:
    sigmaIter+=1
    # approximate flux likelihood, with scaling of both the mean and variance.
    # This approximates the true likelihood with an iterative scheme.
    # - data : fluxes, fluxesVar
    # - model based on SED : f_mod

    b_in = np.array(params['p_t'])[None, :]
    beta2 = np.array(params['p_z_t'])**2.0
    #beta2 = np.zeros(np.array(params['p_z_t']).shape)
    #beta2 += ellPriorSigma**2
    #compute prior on z
    prior = b_in * redshiftGrid[:, None] / beta2[None, :] *np.exp(-0.5 * redshiftGrid[:, None]**2 / beta2[None, :])

    
    #maglim = 26  # M5 magnitude max
    #prior = libPriorPZ(redshiftGrid, maglim=maglim)  # return 2D template nz x nt, nt is 8
    #print(prior.shape)
    
    plotInd = -1
    for ligne in np.arange(4):
        for colonne in np.arange(2):
            plotInd += 1
            axs3[ligne, colonne].plot(redshiftGrid, prior[:, plotInd], label='ellPriorSigma ='+' 1e{}'.format(np.log10(ellPriorSigma)))
            axs3[ligne, colonne].set_xlabel('z-spec')
            axs3[ligne, colonne].set_ylabel('prior')
            axs3[ligne, colonne].set_yscale('log')
            axs3[ligne, colonne].set_title('SED {}'.format(sed_names[plotInd]))
            axs3[ligne, colonne].legend(loc="upper right")

    ellML = approx_flux_likelihood(fluxes, fluxesVar, f_mod, normalized=True, marginalizeEll=True, returnEllML=True, ell_hat=ell_hat_z, ell_var=(ell_hat_z*ellPriorSigma)**2)
    like_grid = approx_flux_likelihood(fluxes, fluxesVar, f_mod, normalized=True, marginalizeEll=True, renormalize=False, ell_hat=ell_hat_z, ell_var=(ell_hat_z*ellPriorSigma)**2)
    #print(like_grid.shape)
            
    plotInd = -1
    for ligne in np.arange(4):
        for colonne in np.arange(2):
            plotInd += 1
            axs4[ligne, colonne].plot(redshiftGrid, like_grid[:, plotInd], label='ellPriorSigma ='+' 1e{}'.format(np.log10(ellPriorSigma)))
            axs4[ligne, colonne].set_xlabel('z-spec')
            axs4[ligne, colonne].set_ylabel('flux Likelihood')
            if np.all(like_grid[:, plotInd] > 0):
                axs4[ligne, colonne].set_yscale('log')
            else:
                axs4[ligne, colonne].set_yscale('linear')
            axs4[ligne, colonne].set_title('SED {}'.format(sed_names[plotInd]))
            axs4[ligne, colonne].legend(loc="upper right")
    
    like_grid_mult = like_grid * prior
    #print(like_grid.shape)
    ## Plot for this iteration on ellPriorSigma:
    alpha = 0.9
    s = 5
    plotInd = -1
    for ligne in np.arange(4):
        for colonne in np.arange(2):
            plotInd += 1
            axs[ligne, colonne].plot(redshiftGrid, ellML[:, plotInd, 0], label='ellPriorSigma ='+' 1e{}'.format(np.log10(ellPriorSigma)))
            axs[ligne, colonne].set_xlabel('z-spec')
            axs[ligne, colonne].set_ylabel('ell ML')
            axs[ligne, colonne].set_yscale('log')
            axs[ligne, colonne].set_title('SED {}'.format(sed_names[plotInd]))
            axs[ligne, colonne].legend(loc="upper right")
            
    plotInd = -1
    for ligne in np.arange(4):
        for colonne in np.arange(2):
            plotInd += 1
            axs2[ligne, colonne].plot(redshiftGrid, like_grid_mult[:, plotInd], label='ellPriorSigma ='+' 1e{}'.format(np.log10(ellPriorSigma)))
            axs2[ligne, colonne].set_xlabel('z-spec')
            axs2[ligne, colonne].set_ylabel('fluxLikelihood * prior')
            if np.all(like_grid_mult[:, plotInd] > 0):
                axs2[ligne, colonne].set_yscale('log')
            else:
                axs2[ligne, colonne].set_yscale('linear')
            axs2[ligne, colonne].set_title('SED {}'.format(sed_names[plotInd]))
            axs2[ligne, colonne].legend(loc="upper right")

    #print("ellML is of dimensions {}".format(ellML.shape))

fig.suptitle('ell ML', fontsize=18,y=1.01)
fig.show()

fig3.suptitle('Prior', fontsize=18,y=1.01)
fig3.show()

fig4.suptitle('Flux likelihood', fontsize=18,y=1.01)
fig4.show()

fig2.suptitle('Flux likelihood * prior', fontsize=18,y=1.01)
fig2.show()
```
